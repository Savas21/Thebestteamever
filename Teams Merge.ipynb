```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
import time
```

```python
#Read CSV files
NA_twitter_data="twitterTEAMS_NA_sentiments.csv"
EU_twitter_data="twitterTEAMS_sentiments_EU.csv"
Win_rate_data="WinLossRates.csv"
```

```python
NA_twitter_df=pd.read_csv(NA_twitter_data,encoding="ISO-8859-1")
NA_twitter_df
```

```python
EU_twitter_df=pd.read_csv(EU_twitter_data,encoding="ISO-8859-1")
EU_twitter_df
```

```python
EU_twitter_df.set_index("Team")
```

```python
## Merge two regions teams data 
team_sentiment_df=pd.merge(NA_twitter_df,EU_twitter_df,how='outer')
team_sentiment_df
```

```python
#Renmae the team names 
team_sentiment_df=team_sentiment_df.replace("@100Thieves","100 Thieves")
team_sentiment_df=team_sentiment_df.replace("@Cloud9","Cloud9")
team_sentiment_df=team_sentiment_df.replace("@ClutchGaming","Clutch Gaming")
team_sentiment_df=team_sentiment_df.replace("@FlyQuestSports","FlyQuest")
team_sentiment_df=team_sentiment_df.replace("@GoldenGuardians","Golden Guardians")
team_sentiment_df=team_sentiment_df.replace("@OpTicGaming","OpTic Gaming")
team_sentiment_df=team_sentiment_df.replace("@TeamLiquid","Team Liquid")
team_sentiment_df=team_sentiment_df.replace("@TeamSoloMid","TSM")
team_sentiment_df=team_sentiment_df.replace("@clgaming","Counter Logic Gaming")
team_sentiment_df=team_sentiment_df.replace("@echofoxgg","Echo Fox")
team_sentiment_df=team_sentiment_df.replace("@Fanatics","Fnatic")
team_sentiment_df=team_sentiment_df.replace("@G2esports","G2 Esports")
team_sentiment_df=team_sentiment_df.replace("@GiantsGamingENG","Giants")
team_sentiment_df=team_sentiment_df.replace("@H2KGG","H2K")
team_sentiment_df=team_sentiment_df.replace("@MisfitsGG","Misfits Gaming")
team_sentiment_df=team_sentiment_df.replace("@ROCCAT","ROCCAT")
team_sentiment_df=team_sentiment_df.replace("@Splyce","Splyce")
team_sentiment_df=team_sentiment_df.replace("@Team_Vitality","Vitality")
team_sentiment_df=team_sentiment_df.replace("@UnicornsOfLove","Unicorns of Love")
team_sentiment_df=team_sentiment_df.replace("@s04","FC Schalke 04 Esports")
team_sentiment_df=team_sentiment_df.set_index("Team")
team_sentiment_df
```

```python
Team_summary=team_sentiment_df.groupby(["Team"]).mean()["Compound Score"]
Team_summary_df=pd.DataFrame(Team_summary)
Team_summary_df.reset_index(level=0,inplace=True)
Team_summary_df
```

```python
Win_rate_df=pd.read_csv(Win_rate_data,encoding="ISO-8859-1")
Win_rate_df
```

```python
Win_rate_df=Win_rate_df.rename(columns={"Unnamed: 0":"Team"})
Win_rate_df
```

```python
## Merge two data  
Team_merge_df=pd.merge(Team_summary_df,Win_rate_df,on='Team')
#team_sentiment_df=team_sentiment_df.set_index("Team")
Team_merge_df
```

```python
win_games=Team_merge_df["Wins"]
lose_games=Team_merge_df["Losses"]
Total_games=win_games+lose_games
Win_rate=win_games/Total_games

Team_merge_df['Win Rate'] = Win_rate
Team_merge_df
```
